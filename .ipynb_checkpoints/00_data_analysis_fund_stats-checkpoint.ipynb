```python
fund_codes = ["KPC", "HKH", "KLU", "KZL", "CPU", "KUT"]
```

```python
# Date kolonunu datetime formatına çevir
df['date'] = pd.to_datetime(df['date'])

# Tarih kolonunu indeks olarak ayarla
df.set_index('date', inplace=True)

# Grafiği çiz
plt.figure(figsize=(15, 10))
for fund in fund_codes:
    fund_data = df[df['code'] == fund]
    plt.plot(fund_data.index, fund_data['price'], label=fund)
plt.legend()
plt.title('Price Trends of Funds')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
# Korelasyon matrisi görselleştirme
for fund in fund_codes:
    plt.figure(figsize=(12, 8))
    fund_data = df[df['code'] == fund]
    correlation_matrix = fund_data[['price', 'market_cap', 'number_of_investors', 'number_of_shares']].corr()

    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title(f'Correlation Heatmap for Fund: {fund}')
    plt.show()
```

```python
df['year'] = df.index.year
df['month'] = df.index.month
df['day_of_week'] = df.index.dayofweek + 1
```

```python
df
```

```python
# Günlük getiriyi hesapla
df['daily_return'] = df.groupby('code')['price'].pct_change()

# Hareketli ortalamalar (MA) ekle
df['MA_7'] = df.groupby('code')['price'].transform(lambda x: x.rolling(window=7).mean())
df['MA_30'] = df.groupby('code')['price'].transform(lambda x: x.rolling(window=30).mean())
```

```python
# RSI hesapla
def calculate_rsi(series, period=14):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

df['RSI'] = df.groupby('code')['price'].transform(lambda x: calculate_rsi(x))
```

```python
df
```

```python
# RSI'yi çizdirmek için
def plot_rsi(df, period=14):
    plt.figure(figsize=(14, 7))

    # RSI grafiği
    plt.subplot(2, 1, 1)
    plt.plot(df.index, df['price'], label='Price')
    plt.title('Price')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()

    plt.subplot(2, 1, 2)
    plt.plot(df.index, df['RSI'], label=f'RSI ({period} days)', color='orange')
    plt.axhline(70, color='red', linestyle='--', label='Overbought (70)')
    plt.axhline(30, color='green', linestyle='--', label='Oversold (30)')
    plt.title('RSI')
    plt.xlabel('Date')
    plt.ylabel('RSI')
    plt.legend()

    plt.tight_layout()
    plt.show()

# RSI grafiğini çiz
plot_rsi(df)
```
